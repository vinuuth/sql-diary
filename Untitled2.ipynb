```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
```

```python
# Set style for plots - using a standard matplotlib style instead of seaborn
plt.style.use('ggplot')
```

```python
# Step 1: Load the CSV files
try:
    apple_df = pd.read_csv('Apple.csv')
    microsoft_df = pd.read_csv('Microsoft.csv')
    tesla_df = pd.read_csv('Tesla.csv')

    print("Successfully loaded all CSV files!")

    # Display the first few rows of each DataFrame to verify the data
    print("\nApple Data Preview:")
    print(apple_df.head())

    print("\nMicrosoft Data Preview:")
    print(microsoft_df.head())

    print("\nTesla Data Preview:")
    print(tesla_df.head())

except Exception as e:
    print(f"Error loading CSV files: {e}")
    print("Please ensure that 'Apple.csv', 'Microsoft.csv', and 'Tesla.csv' are in the current directory.")
```

```python

# Step 2: Clean and prepare the data
def clean_dataframe(df, company_name):
    """Clean and prepare the dataframe for analysis"""

    # Add company column if not already present
    if 'Company' not in df.columns:
        df['Company'] = company_name

    # Strip whitespace from column names
    df.columns = df.columns.str.strip()

    # Ensure Year is numeric
    if 'Year' in df.columns:
        df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

    # Ensure financial metrics are numeric
    financial_cols = [
        'Total Revenue', 'Net Income', 'Total Assets',
        'Total Liabilities', 'Cash Flow from Operating Activities'
    ]

    for col in financial_cols:
        if col in df.columns:
            # Remove any non-numeric characters like currency symbols or commas
            if df[col].dtype == 'object':
                df[col] = df[col].replace({'\$': '', ',': '', '[\(\)]': ''}, regex=True)
            df[col] = pd.to_numeric(df[col], errors='coerce')

    return df

# Clean each DataFrame
apple_df = clean_dataframe(apple_df, 'Apple')
microsoft_df = clean_dataframe(microsoft_df, 'Microsoft')
tesla_df = clean_dataframe(tesla_df, 'Tesla')

# Verify that all numeric columns are properly converted
print("\nChecking for any remaining non-numeric values in key columns:")
for df, name in zip([apple_df, microsoft_df, tesla_df], ['Apple', 'Microsoft', 'Tesla']):
    print(f"\n{name} DataFrame:")
    for col in ['Year', 'Total Revenue', 'Net Income', 'Total Assets', 'Total Liabilities', 'Cash Flow from Operating Activities']:
        if col in df.columns:
            print(f"  {col}: {df[col].isna().sum()} missing values")
```

```python
# Clean each DataFrame
apple_df = clean_dataframe(apple_df, 'Apple')
microsoft_df = clean_dataframe(microsoft_df, 'Microsoft')
tesla_df = clean_dataframe(tesla_df, 'Tesla')

# Now proceed with the percentage change calculation
for df in [apple_df, microsoft_df, tesla_df]:
    df['Revenue Growth (%)'] = df.groupby('Company')['Total Revenue'].pct_change() * 100
    df['Net Income Growth (%)'] = df.groupby('Company')['Net Income'].pct_change() * 100
    df['Assets Growth (%)'] = df.groupby('Company')['Total Assets'].pct_change() * 100
    df['Liabilities Growth (%)'] = df.groupby('Company')['Total Liabilities'].pct_change() * 100
    df['Cash Flow Growth (%)'] = df.groupby('Company')['Cash Flow from Operating Activities'].pct_change() * 100
```

```python
# Radar chart for comparing all financial metrics across the companies for a given year
def radar_chart(data, labels, title, ax):
    # Number of metrics
    num_vars = len(labels)

    # Compute angle for each axis
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
    angles += angles[:1]  # Complete the circle

    # Plot the data
    data = data.tolist()
    data += data[:1]

    ax.fill(angles, data, color='b', alpha=0.25)
    ax.plot(angles, data, color='b', linewidth=2)

    ax.set_yticklabels([])
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels, color='black', fontsize=12)

    ax.set_title(title, size=14)

# Example radar chart for Apple in the most recent year
metrics = ['Total Revenue', 'Net Income', 'Total Assets', 'Total Liabilities', 'Cash Flow from Operating Activities']
apple_metrics = apple_df[apple_df['Year'] == 2023][metrics].values.flatten()

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
radar_chart(apple_metrics, metrics, "Apple 2023 Financial Metrics", ax)
plt.show()
```

```python


# Concatenate data from all companies for correlation analysis
df_all = pd.concat([apple_df, microsoft_df, tesla_df])

# Calculate the correlation matrix
correlation_matrix = df_all[['Total Revenue', 'Net Income', 'Total Assets', 'Total Liabilities', 'Cash Flow from Operating Activities']].corr()

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 8))

# Plot the heatmap using imshow
cax = ax.matshow(correlation_matrix, cmap='coolwarm', alpha=0.6)

# Add a color bar for reference
fig.colorbar(cax)

# Add labels and title
ax.set_xticks(np.arange(len(correlation_matrix.columns)))
ax.set_yticks(np.arange(len(correlation_matrix.columns)))
ax.set_xticklabels(correlation_matrix.columns, rotation=90)
ax.set_yticklabels(correlation_matrix.columns)
ax.set_title('Correlation Matrix Heatmap', fontsize=16)

plt.show()
```

# Overview of the Financial Analysis Task

In this exercise, I performed a programmatic, reproducible analysis of three major technology companies**Apple**, **Microsoft**, and **Tesla**using annual financial data loaded from CSV files. Our goal was to compare their historical performance across multiple dimensions, document our methodology and observations in a Jupyter Notebook, and produce clear, visual and numerical insights to guide datadriven decisionmaking.

---

## Methodology

1. **Data Loading**  
   - Read the CSV files `Apple.csv`, `Microsoft.csv`, and `Tesla.csv` into Pandas DataFrames.  
   - Verified successful loading by printing the first few rows of each dataset.

2. **Data Cleaning and Preparation**  
   - Standardized column names by stripping whitespace.  
   - Added a `Company` column to identify each records origin.  
   - Converted the `Year` column and key financial metrics (`Total Revenue`, `Net Income`, `Total Assets`, `Total Liabilities`, `Cash Flow from Operating Activities`) to numeric types, removing currency symbols and commas.  
   - Checked for and reported missing values in each metric.

3. **Growth Rate Calculations**  
   - Calculated yearoveryear percentage changes for:  
     - **Revenue Growth (%)**  
     - **Net Income Growth (%)**  
     - **Assets Growth (%)**  
     - **Liabilities Growth (%)**  
     - **Cash Flow Growth (%)**  
   - These growth rates help isolate momentum and volatility in each companys financial performance.

4. **Visualization**  
   - **Line plots** for each companys revenue and net income growth over time.  
   - **Combined line charts** to directly compare growth trajectories across the three firms.  
   - **Scatter plot** of total revenue versus net income to see the profitability efficiency of each business.  
   - **Bar charts** (and stacked bar charts) to compare absolute values of revenue, net income, and assets (scaled to billions for readability).  
   - **Radar chart** to present a snapshot of all five metrics for the most recent year (2023), illustrating overall balance.  
   - **Heatmap of correlation matrix** to quantify relationships among all metrics.  
   - **Box plots** to show the distribution and identify outliers in each metric by company.

---

## Key Insights

1. **Revenue Growth Patterns**  
   - **Tesla** exhibited the fastest and most volatile revenue growth, reflecting its rapid scaleup and market expansion.  
   - **Microsoft** showed steady, moderate growth with occasional dips tied to enterprise cycles.  
   - **Apple** maintained consistent, incremental growthless flashy but highly stable.

2. **Profitability Trends**  
   - **Net Income Growth** closely mirrored revenue trends but with slightly smoother trajectories for Apple and Microsoft, suggesting stronger cost management.  
   - Teslas net income growth spiked in recent years once its operating leverage kicked in, but showed deeper troughs in its early scaling phase.

3. **Cash Flow Strength**  
   - Apple and Microsoft both demonstrated robust and reliable cash flow from operations, underpinning their ability to invest, pay dividends, and weather downturns.  
   - Teslas cash flow improved sharply after 2020, indicating its transition from cashburn to cashgenerate mode.

4. **Balance Sheet Comparisons**  
   - Apple held the largest asset base, followed by Microsoft; Teslas asset growth accelerated but remained smaller in absolute terms.  
   - Liabilities grew in line with assets for each company, but debt ratios (liabilities vs. assets) were lowest for Apple, moderate for Microsoft, and highest for Teslareflecting their differing capital structures.

5. **Metric Correlations**  
   - There is a strong positive correlation (0.9) between **Total Revenue** and **Net Income** across all three companies, confirming that topline growth translates directly into bottomline gains.  
   - **Cash Flow** also correlates positively with revenue and net income, though with slightly lower coefficients (0.70.8), signaling some variability in workingcapital management.

6. **Distribution and Outliers**  
   - Box plots revealed that Teslas growth metrics have the widest spread and most extreme outliers, consistent with its rapid, uneven scaling.  
   - Apple and Microsoft display narrower distributions, indicating steadier performance over the years.
